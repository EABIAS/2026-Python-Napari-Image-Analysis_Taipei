# Part I. 3D volume visualization

這一部分先熟悉 napari 在 3D 影像中的基本視覺化概念。

## Concept
**Volume rendering** 是將整個 voxel intensity 分布以 3D 方式呈現。  
它很適合回答：

- 訊號大致分布在哪裡？
- 深度方向上的亮度如何變化？
- 多通道訊號在空間中如何重疊？

但它 **不等於真正把物體表面萃取出來**。


## 1. Open a built-in 3D sample

這裡使用 napari 內建的 `cells3d` 範例資料。


```python

viewer.layers.clear()
viewer.open_sample('napari', 'cells3d')
viewer.dims.ndisplay = 2
viewer
```

### Try it yourself

2. 再切換到 3D 模式 (`viewer.dims.ndisplay = 3`)  
3. 比較 2D slice 與 3D volume 的資訊差異


```python

viewer.dims.ndisplay = 3
viewer
```

## 2. Inspect image-layer rendering parameters

在 3D 模式下，Image layer 常調整的參數包括：

- `opacity`
- `blending`
- `contrast_limits`
- `rendering`
- `interpolation`

### 常見 rendering 模式
- `mip`：maximum intensity projection
- `attenuated_mip`：帶有深度衰減感
- `iso`：呈現等值面風格


```python

# 觀察目前有哪些 layer
[layer.name for layer in viewer.layers]
```

```python

# 依照實際 layer 名稱調整參數（cells3d 通常會分成兩個 channel）
for layer in viewer.layers:
    if isinstance(layer, napari.layers.Image):
        layer.rendering = 'attenuated_mip'
        layer.interpolation = 'linear'
        layer.opacity = 0.7

viewer
```

## 3. Create a synthetic 5D demo dataset

這一段對應課程中 `(T, C, Z, Y, X)` 的 5D 範例。  
重點是：

- `channel_axis=1`
- 不同 channel 指定不同 colormap
- 設定 `axis_labels`


```python

viewer.layers.clear()

data = np.random.random((5, 3, 20, 256, 256))

viewer.add_image(
    data,
    name='5D_demo',
    channel_axis=1,
    colormap=['red', 'green', 'blue'],
    blending='additive'
)

viewer.dims.axis_labels = ('T', 'C', 'Z', 'Y', 'X')
viewer.dims.ndisplay = 2
viewer
```

```python

viewer.dims.ndisplay = 3

for layer in viewer.layers:
    layer.rendering = 'attenuated_mip'
    layer.interpolation = 'linear'
    layer.opacity = 0.6

viewer
```

### Discussion

在這個 5D dataset 中：

- `T` = time
- `C` = channel
- `Z` = depth
- `Y`, `X` = spatial axes

這種資料格式在顯微影像中特別常見。


# Part II. From volume to surface / mesh

這一部分要回答的問題是：

> 看到 3D volume，和真正抽出物體表面，差別在哪裡？

## Key message

- **Volume rendering**：看到的是訊號分布
- **Surface extraction**：得到的是物體邊界的幾何表示


## 4. Create a solid sphere volume

先自己建立一個 3D 球體，讓 surface extraction 的概念更清楚。


```python

viewer.layers.clear()

Z, Y, X = 64, 128, 128
zz, yy, xx = np.meshgrid(
    np.arange(Z),
    np.arange(Y),
    np.arange(X),
    indexing='ij'
)

cz, cy, cx = 32, 64, 64
r = 30

sphere = ((zz - cz) ** 2 + (yy - cy) ** 2 + (xx - cx) ** 2) < r ** 2

viewer.add_image(sphere.astype(float), name='sphere')
viewer.dims.axis_labels = ('Z', 'Y', 'X')
viewer.dims.ndisplay = 3
viewer.layers['sphere'].rendering = 'iso'
viewer
```

## 5. Compare rendering modes

請分別試試看：

- `mip`
- `attenuated_mip`
- `iso`

觀察它們呈現球體的方式有什麼不同。


```python

layer = viewer.layers['sphere']

# 你可以反覆修改這一行：'mip' / 'attenuated_mip' / 'iso'
layer.rendering = 'attenuated_mip'
layer.opacity = 0.8
viewer
```

## 6. Extract a real surface mesh with marching cubes

這一步非常重要。  
`marching_cubes` 會從 volumetric data 中抽出等值面，產生：

- `verts`：mesh 頂點座標
- `faces`：三角形連接關係

這樣得到的才是真正可用於幾何分析的 surface mesh。


```python

verts, faces, normals, values = marching_cubes(sphere.astype(np.uint8), level=0.5)

print("verts shape:", verts.shape)
print("faces shape:", faces.shape)
```

```python

viewer.add_surface((verts, faces), name='sphere_surface')
viewer
```

## 7. Why surface matters

拿到 surface mesh 之後，常見可進一步分析的量包括：

- surface area
- enclosed volume
- shape descriptor
- curvature
- mesh-based animation / modeling

也就是說：

- volume rendering 適合看訊號分布
- surface mesh 適合看邊界、形狀與量化


## 8. Optional: build a more irregular 3D object

若想讓示範更像真實生物影像，可試試不規則 blob。


```python

viewer.layers.clear()

blob = binary_blobs(length=96, n_dim=3, volume_fraction=0.12).astype(np.uint8)
viewer.add_image(blob.astype(float), name='blob')
viewer.dims.axis_labels = ('Z', 'Y', 'X')
viewer.dims.ndisplay = 3
viewer.layers['blob'].rendering = 'iso'

verts_blob, faces_blob, _, _ = marching_cubes(blob, level=0.5)
viewer.add_surface((verts_blob, faces_blob), name='blob_surface')

viewer
```

# Part III. Scientific animation with napari-animation

這部分主要是 **workflow concept**。  
實際輸出影片時，最方便的方式通常還是透過 napari GUI 的 Animation Wizard。

## Key concepts

### Keyframe
手動指定的重要畫面，例如：
- 全景
- 放大某個細胞
- 旋轉到側面
- 再 zoom in 到 nucleus

### Interpolation
在兩個 keyframe 之間自動補中間畫面。

### Easing
控制畫面移動速度的變化方式，例如：
- 緩慢起步
- 中間較快
- 結尾再慢下來


## 9. Recommended GUI workflow for napari-animation

在 napari 中可依序操作：

1. 開啟影像
2. 切到想要的視角 / zoom / time point
3. 開啟 `Tools -> Visualization -> Animation Wizard`
4. Add keyframe
5. 再改變 camera 或 layer state
6. Add next keyframe
7. 設定 `steps`, `FPS`, `quality`
8. Render animation

### Practical interpretation
動畫長度約為：

`總 steps ÷ FPS`

例如：
- `steps = 60`
- `FPS = 30`

則影片長度約 2 秒。


## 10. A small helper cell for planning animation timing

這個 cell 不會直接輸出動畫，但可幫助你規劃教學時的設定。


```python

def estimate_duration(total_steps, fps):
    return total_steps / fps

examples = pd.DataFrame({
    'total_steps': [30, 60, 90, 120],
    'fps': [15, 30, 30, 24]
})
examples['duration_sec'] = examples.apply(lambda r: estimate_duration(r['total_steps'], r['fps']), axis=1)
examples
```

## 11. Why use napari-animation?

相較於只能手動錄影或單純截圖的方式，napari-animation 的優點在於：

- keyframe-based workflow
- 可重複調整 camera path
- 容易整合 layer state 的變化
- 適合 scientific presentation 與 teaching demonstration
- 更容易建立可重現的展示流程

## 12. Create a simple 3D tracks example

Tracks layer 的基本欄位格式為：

`track_id, time, z, y, x`


```python

viewer.layers.clear()

tracks = np.array([
    [0, 0, 10, 120,  80],
    [0, 1, 11, 123,  85],
    [0, 2, 12, 128,  92],
    [1, 0, 18, 200, 150],
    [1, 1, 18, 205, 155],
    [1, 2, 19, 210, 160],
], dtype=float)

viewer.add_tracks(tracks, name='tracks_3d')
viewer.dims.axis_labels = ('T', 'Z', 'Y', 'X')
viewer.dims.ndisplay = 3
viewer
```

## 13. Add a matching dummy image under the tracks

有底圖時，tracking 的示意通常更清楚。


```python

viewer.layers.clear()

dummy = np.zeros((3, 32, 256, 256), dtype=np.float32)

# 在 track 附近放一些亮點，方便視覺化
for row in tracks.astype(int):
    track_id, t, z, y, x = row
    z = np.clip(z, 0, dummy.shape[1] - 1)
    y = np.clip(y, 0, dummy.shape[2] - 1)
    x = np.clip(x, 0, dummy.shape[3] - 1)
    dummy[t, z, max(0, y-2):y+3, max(0, x-2):x+3] = 1

viewer.add_image(dummy, name='dummy_image')
viewer.add_tracks(tracks, name='tracks_3d')
viewer.dims.axis_labels = ('T', 'Z', 'Y', 'X')
viewer.dims.ndisplay = 3
viewer
```

## 14. Import tracking data from CSV

如果你的追蹤結果已經存在 CSV，可用下列方式讀進 napari。


```python

# 範例：先把 tracks 寫成 csv，再讀回來
df_tracks = pd.DataFrame(tracks, columns=['track_id', 't', 'z', 'y', 'x'])
df_tracks.to_csv('tracks_3d_example.csv', index=False, encoding='utf_8_sig')
df_tracks.head()
```

```python

df = pd.read_csv('tracks_3d_example.csv')
tracks_from_csv = df[['track_id', 't', 'z', 'y', 'x']].to_numpy(dtype=float)

viewer.layers.clear()
viewer.add_tracks(tracks_from_csv, name='tracks_from_csv')
viewer.dims.axis_labels = ('T', 'Z', 'Y', 'X')
viewer.dims.ndisplay = 3
viewer
```

## 15. Basic quantitative summaries from tracks

napari 的 Tracks layer 偏向視覺化，但表格資料本身仍可做基本量化，例如：

- 每條 track 的 frame 數
- 起點與終點
- 位移距離


```python

df = pd.DataFrame(tracks_from_csv, columns=['track_id', 't', 'z', 'y', 'x'])

summary = (
    df.groupby('track_id')
      .agg(
          n_points=('t', 'size'),
          t_start=('t', 'min'),
          t_end=('t', 'max'),
          z_start=('z', 'first'),
          y_start=('y', 'first'),
          x_start=('x', 'first'),
          z_end=('z', 'last'),
          y_end=('y', 'last'),
          x_end=('x', 'last'),
      )
      .reset_index()
)

summary['displacement'] = np.sqrt(
    (summary['z_end'] - summary['z_start']) ** 2 +
    (summary['y_end'] - summary['y_start']) ** 2 +
    (summary['x_end'] - summary['x_start']) ** 2
)

summary
```

## 18. Inspect multiscale pyramid levels

如果某個 layer 是 multiscale，可以列出每一層的 shape。


```python

if len(viewer.layers) > 0:
    layer = viewer.layers[0]
    if getattr(layer, 'multiscale', False):
        print("Multiscale pyramid levels:")
        for i, arr in enumerate(layer.data):
            print(f"Level {i}: shape = {arr.shape}")
    else:
        print("This layer is not multiscale.")
        if hasattr(layer.data, 'shape'):
            print("Shape =", layer.data.shape)
else:
    print("No layer found. Please open a dataset first.")
```

## 19. Understanding multiscale behavior

典型情況下：

- **Level 0**：full resolution
- **Level 1**：較低解析度
- **Level 2**：更低解析度

當你看全景時，viewer 先讀低解析度版本；  
放大到局部時，再逐步讀更高解析度資料。

這和地圖服務的縮放邏輯很像。


## 20. Why OME-Zarr matters

OME-Zarr 的價值不只是「檔案很大也能存」而已，而是它讓下面幾件事更實用：

- 大型影像可分塊存取
- 僅載入當前視野所需資料
- 多尺度瀏覽更順暢
- 更適合遠端 / 雲端資料分享
- 更容易支援可重現的分析流程
