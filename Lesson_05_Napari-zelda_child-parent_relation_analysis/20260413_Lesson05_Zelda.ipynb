### 1️⃣ Activate napari

```python
from skimage.io import imread
import napari
viewer = napari.Viewer() #  啟動napari視窗
```

### 2️⃣ Activate plugin 'napari-zelda'

```python
from qtpy.QtCore import QTimer

def run_zelda():
    zelda_gui.dropdown.value = "Segment two object populations and relate Parent to Child"
    zelda_gui.call_button.click()

dock, zelda_gui = viewer.window.add_plugin_dock_widget('napari-zelda')
QTimer.singleShot(500, run_zelda)

zelda_gui.dropdown.value = "Segment two populations and relate"
zelda_gui()
```

### 3️⃣ Open raw image, parent labels, child labels

```python
fp_DNA_lab=r"./zelda_label_batch/0_p.tif"
fp_RNA_lab=r"./zelda_label_batch/0_c.tif"
fp_RNA_raw=r"./zelda_label_batch/RNA.tif"

# Add image to viewer 
data = [imread(fp_RNA_raw), imread(fp_DNA_lab), imread(fp_RNA_lab)]
viewer.add_image(data[0], name="RNA_raw", colormap='gray')
viewer.add_image(data[1], name="parents_lab", colormap='inferno')
viewer.add_image(data[2], name="children_lab", colormap='inferno')
```
