## **2.1 打開分析工具箱：套件、模組與 API**  
在 Python 中工具的種類五花八門，而這些工具被有系統地整理起來並依照功能與用途，被分成不同層級：  
* **套件(Package):**   
  一整個工具家族（例如：`skimage`）。   
   
   
* **模組 (Module)：**  
  套件中的功能分類，負責特定的任務（例如：`filters` 專門處理濾波任務）。  


* **API / 函式 (Function)：**  
  模組中真正執行工作的工具（例如：`filters.gaussian`）。


* **參數 (Arguments)：**  
  使用工具時的「細部設定」（例如：妳要多強的濾波？`sigma=2`）。  
>**你可以這樣想像:**   
> 套件是整座五金行，模組是工具櫃（如電鑽櫃），API 是電鑽，參數是你使用電鑽時選用的轉速與鑽頭。  

當我們知道工具是如何分層的，下一步要關心的，就是：  
**這些工具在使用時，程式碼通常長什麼樣子？**

---
## **2.2 API 的通用結構**  
不論是影像處理、讀取或視覺化工具，多數的 API 都遵循同一個基本模式: 
```PYthon

output = api_function(input, parameter1=value1, parameter2=value2...)

```
>這行程式碼可以這樣理解 : 「把 input 丟進 api_function，用指定的參數處理後，把結果存成 output。」  

這種 **「 給它資料 -> 它處理 -> 回傳結果」** 就是多數 API 在做的事情

我們在操作 API 時，要清楚:  
- 輸入(input)  
- 輸出(output)   
- 必要參數(arguments) 

才能把不同的 API 像積木一樣拼接在一起。

## **2.3 API 載入與呼叫技巧(Import & Call)** 
掌握了結構後，我們要學習如何正確「正確招喚」它們，避免發生 「找不到工具」的窘境。  

一個重要的原則是:
>**你載入到哪一層，呼叫時就只能從那一層往下寫。**

---
#### **載入方式，會直接影響呼叫的寫法**  
考慮到代碼的可讀性，建議根據工具的用途來選擇載入方式，而常有的呼叫方式有以下幾種:  
* **載入整個「套件」(不推薦，太累):**  
```Python
import skimage
```
>這表示我們把整個套件(Package)載入進來，呼叫時**要寫完整的工具路徑:**    
```Python
skimage.filters.gaussian()
```

* **載入到 「子模組」 (最推薦):**   
```Python
from skimage import filters 
```
>這表示我們把整類工具(filters) 載入進來，而呼叫時**必須保留模組名稱:** 
    
```Python
filters.gaussian(image, sigma=2)
```
    優點是呼叫時帶上分類標籤，既能保持邏輯清晰又方便查詢，程式碼可讀性更高。
---       
* **載入到「方法」 (適合高頻率工具)：**   
```Python
from skimage.filters import gaussian 
```  
>這表示我們直接載入單一 API，呼叫時可以**直接使用函式名稱：**
    
```Python
 gaussian(image, sigma=2)
```
    這種呼叫最精簡，適合幾乎每段程式碼都會用到的基本功能。

## **2.4 常用的影像處理 API**   
掌握正確的招喚技巧後，下面的列表是生物影像分析最常用的工具:  

| 工具 / 模組 | 功能定位 | 常見用途（生物影像語言） | 範例 API |
|:-----------|:---------|:------------------------|:---------|
| skimage | 影像讀取 | 載入標準格式影像 (TIFF, PNG, JPG) 到 NumPy Array | `imread` |
| bioio | 影像讀取 | 讀取顯微鏡原廠格式 (如 .czi, .nd2, .lif) 並自動解析維度訊息 | `BioImage` |
| skimage | 影像處理與分析 | 影像平滑化與去雜訊 | `filters.gaussian`、`filter.median`|
| skimage | 影像分割 | 自動計算背景門檻值，將訊號與背景分離（產生遮罩） | `filters.threshold_otsu`、`filters.threshold_local` |
| skimage | 形態學處理 | 針對「二值化遮罩」（數值 0/1 或 True/False）進行修整。修補斷裂、填洞、去雜訊 | `morphology.opening` |
| skimage | 區域量測 | 特徵定量分析：計算單個細胞的面積、圓形度、平均螢光強度等 | `measure.regionprops` |
| stackview | 影像檢視 | 互動式檢查：瀏覽 Z-stack / 多通道影像 | `stackview.StackView` |
| matplotlib.pyplot | 視覺化 | 顯示影像、疊加結果、畫圖 | `plt.imshow` |
