# Part 1: 影像的數位基因 - 變數與像素 (Variables & Dtypes)
### **1.1 影像的本質就是數值**
我們看到的影像，其實是由一層層的資訊堆疊來的...

**Layer 1: 整張影像(The whole image)**  
在 Python 裡，我們會把它存進一個**變數(Variables)**，例如 `rawImg` ，後續所有操作，都是對這個變數進行。

**Layer 2: 像素(Pixel)**  
影像是由許多小格子組成，每一個小格子就是一個像素，也是影像的最小單位。。  
      
**Layer 3: 數值(The value)**  
    每一個像素，對應到一個數值，代表亮度或強度。 
    
#### 在電腦眼中，影像就是一個由數值組成的矩陣；而影像分析，就是對這個矩陣進行操作。

#### **1.2.2 變數的使用規則：資料型別(Data Types)**
資料型別，決定了這個變數 [裝的是什麼]，以及 [你可以對它做什麼]!!  

#### 在影像分析中，最常見的資料型別有以下幾種:  
----
**1. Integer(整數):像素原始值** 
* uint8 / uint16 : 原始影像最常用的像素值(Pixel value)  
* 用途: 儲存原始影像的像素值（Pixel value），為了節省空間，它只存整數。
* uint8: 8-bit image，數值從 0 ~ 255；uint16: 16-bit image，數值從 0 ~ 65535
---
    
**2.  Float(浮點數):計算用數值**  
* 當影像經過運算（如：除以、Gaussian Blur）產生小數點時，資料會自動轉型成 float  
* 用途：這是影像分析理想的 **計算格式**。
* Python 的計算預設是 float64，以維持數值模擬、統計獲最佳化問題需要的高精度  　　
* 影像分析更常用的是 float32 
---    

**3. Boolean(布林值):判斷與遮罩**  
* 布林值只會有兩種狀態，用來表示「真 / 假」的邏輯判斷結果。  
* 真：`True`(1)；假：`False`(0) 
* 用途：表示條件判斷的結果
* 在影像分析中，常用來表示 遮罩（Mask）
* **影像分割（Segmentation）的結果，本質上就是 Boolean mask。**
---

* 用來表示文字或路徑。  
* 用 引號(`' '`或 `" "`) 包起來的文字或數字。

    3. 類別或條件名稱: 例如細胞種類

#### **1.3.2 從 List 取出資料：索引與輪迴**
>這一節的目的，不是要你學會寫程式語法，  
>而是讓你知道：  
>**放進 List 裡的資料，可以被「指定取出」，也可以被「依序處理」。**

---  
#### **一、用「索引」提取 list 内的資料**  
List 裡的資料是有**順序**的，我們可以透過索引值（Index) 來指定要取出哪一個資料。
概念上，可以這樣理解:  
* List[index] ->取出第 index +1 的資料
* **注意第一筆資料的索引值為 0 而不是 1**



````python
#舉例來說，當我們想取 images list 裡第一個資料
NameOfImages = ["img1", "img2", "img3"]
print(NameOfImages[0])
````

> **補充：索引也可以從後面開始**  
> 除了從前面取值之外，我們也可以使用「負索引」從後面取值：  
> List[-1] -> 取出倒數最後一個元素  
> List[-2] -> 取出倒數第二個元素  

#### **二、用「輪迴(for loop)」依序處理List裡所有的資料**  
在實際的影像分析中，我們通常要處理的是一整群資料，譬如:  
>* 多張影像
>* 多個檔案路徑
>* 多個分析結果

面對這種情況，我們會希望能有一段程式碼可以反覆的執行，並套用到整群資料上。  
這時候就推薦 List 搭配 **for迴圈** 來進行。  

---
迴圈的邏輯非常直覺，她就像是：「針對這疊資料裡的每一份子，都執行相同的動作。」，在影像分析裡面最常見的寫法是:  
**Item-based for loop(直接拿取物件):**
```Python
for i in List:
    # 在這個縮排區塊內，i 代表「目前手上拿的那個物件」
    print(i)
    # 縮排以內都是你會對 List 內每一個物件要重複做的事情
```


````python
#舉例來說，我們當想要查看 某個 List 底下所有資料的訊息:
for img in NameOfImages:
    print(img)
print("迴圈結束惹!!!")
````

**補充說明-(Index-based for loop)**
>另一種迴圈的寫法是**拿取索引序號 (Index-based)**，這在進階課程或實務腳本中很常出現!!
> 語法: 
>```Python
>for i in range(起始值, 結束值):
>   #i 代表索引 (0, 1, 2...)
>   # range()函數代表數字範圍，它會產生一個數字序列：該序列的起點為 啟始值，結束點為 結束值-1  
>   # 當迴圈被執行時，該數列内的每一個數字都會依照順序被迭代一次   
>   # 縮排以內都是每一次迭代要重複做的事情
>   print(i)
>```
